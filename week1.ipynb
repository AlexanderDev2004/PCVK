D2. Pengenalan Python untuk membaca dan menampilkan image dengan
pengenalan macam-macam Library (Numpy, pandas, cv2, skimage, PIL dan
matplotlib)

````python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pyplot as plt
````

````python
# membuat list untuk menyimpan url dari beberapa images

urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]

for url in urls :
  try:
    Image = io.imread(url)
    Image = cv.resize(Image, (0,0), fx=0.5, fy=0.5 )
    Image_2 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((Image, Image_2))
    cv2_imshow(final_frame)
    print('\n')
  except Exception as e:
    print(f"Error processing {url}: {e}")
````

````python
tinggi = Image_2.shape[0]
lebar = Image_2.shape[1]
print ("Resolusi images : Tinggi x lebar = ", tinggi, "X", lebar)
cv2_imshow(Image_2)
````

````python
Image_2 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)
Image_3 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)


for y in range (lebar) :
  Image_3[int(tinggi/2), y] = [255,255,255]

final_frame = cv.hconcat((Image_2, Image_3))
cv2_imshow(Image_3)
````

````python
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"]
for url in urls :
  try:
    Image = io.imread(url)
    Image = cv.resize(Image, (0,0), fx=0.5, fy=0.5 )
    Image_2 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)
    final_frame = cv.hconcat((Image, Image_2))
    cv2_imshow(final_frame)
    print('\n')
  except Exception as e:
    print(f"Error processing {url}: {e}")
````

````python
tinggi = Image_2.shape[0]
lebar = Image_2.shape[1]
print ("Resolusi images : Tinggi x lebar = ", tinggi, "X", lebar)
cv2_imshow(Image_2)
````

2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

````python
# Konversi ke RGB
Image_2 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)

# Copy untuk versi modifikasi
Image_vertikal = Image_2.copy()
Image_diagonal = Image_2.copy()

tinggi, lebar, _ = Image.shape

# --- Versi 1: Garis Vertikal ---
for y in range(tinggi):
    Image_vertikal[y, lebar // 2] = [255, 255, 255]  # garis putih di tengah vertikal

# --- Versi 2: Garis Diagonal ---
for i in range(min(tinggi, lebar)):
    Image_diagonal[i, i] = [255, 255, 255]  # diagonal utama (↘)
    Image_diagonal[i, lebar - i - 1] = [255, 255, 255]  # diagonal lain (↙) → jadi silang (X)

# Gabungkan untuk dibandingkan
final_frame = cv.hconcat((Image_vertikal, Image_diagonal))

# Tampilkan
cv2_imshow(final_frame)
````

3. buat garis horisontal berwarna putih dibagian tengah gambar dengan
panjang tertentu

````python
# Konversi ke RGB
Image_2 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)

# Copy untuk versi garis horizontal
Image_horizontal = Image_2.copy()

tinggi, lebar, _ = Image.shape

# --- Garis Horizontal di Tengah ---
y = tinggi // 2  # posisi garis di tengah vertikal

# Tentukan panjang garis (misalnya 50% dari lebar gambar)
panjang = lebar // 2

# Hitung titik awal & akhir (agar simetris di tengah)
start_x = (lebar - panjang) // 2
end_x = start_x + panjang

# Warnai putih di baris tengah
for x in range(start_x, end_x):
    Image_horizontal[y, x] = [255, 255, 255]

# Tampilkan hasil
cv2_imshow(Image_horizontal)
````

4. Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat
dalam gambar

````python
# Konversi ke RGB
Image_2 = cv.cvtColor(Image, cv.COLOR_BGR2RGB)

# Copy untuk versi kotak
Image_box = Image_2.copy()

tinggi, lebar, _ = Image.shape

# --- Tentukan ukuran kotak ---
width_box = 80   # lebar kotak
height_box = 60  # tinggi kotak

# --- Tentukan posisi acak (pojok kiri atas) ---
import random # Import the random module
x_start = random.randint(0, lebar - width_box - 1)
y_start = random.randint(0, tinggi - height_box - 1)

# --- Gambar kotak putih ---
for y in range(y_start, y_start + height_box):
    for x in range(x_start, x_start + width_box):
        Image_box[y, x] = [255, 255, 255]

# Tampilkan hasil
cv2_imshow(Image_box)
````
